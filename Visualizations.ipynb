# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

# Load the dataset

```python
BAC = pd.read_csv('/content/BAC.xls')
```

# Summary of the datasets

```python
BAC.info()
```

# Data Transformation

```python
BAC['Date']=pd.to_datetime(BAC['Date'])
```

```python
BAC.info()
```

```python
BAC.describe()
```

# Visualization

# Line Plot

```python
# Plotting with Matplotlib

# Line Plot
plt.figure(figsize=(12, 6))
plt.plot(BAC['Date'], BAC['Close'])
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

# Bar Plot


```python
plt.figure(figsize=(12, 6))
plt.bar(BAC['Date'], BAC['Volume'])
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.show()
```

# Histogram


```python
plt.figure(figsize=(12, 6))
plt.hist(BAC['Close'], bins=50)
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

# Scatter Plot


```python
plt.figure(figsize=(12, 6))
plt.scatter(BAC['Open'], BAC['Close'])
plt.title('Opening vs Closing Prices')
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.show()
```

# Plotting with Seaborn



# Line Plot


```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=BAC)
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

# Plotting with Plotly



# Line Plot


```python
fig = px.line(BAC, x='Date', y='Close', title='Stock Prices Over Time')
fig.show()
```

```python
# Bar Plot
fig = px.bar(BAC, x='Date', y='Volume', title='Trading Volume Over Time')
fig.show()
```

```python
# Histogram
fig = px.histogram(BAC, x='Close', nbins=50, title='Distribution of Closing Prices')
fig.show()
```

```python
# Scatter Plot
fig = px.scatter(BAC, x='Open', y='Close', title='Opening vs Closing Prices')
fig.show()
```
